```python
import pandas as pd
import numpy as np
from datetime import datetime
df = pd.read_csv('curah_hujan_new.csv',delimiter=";",decimal=",")
df["Tanggal"] = pd.to_datetime(df["Tanggal"])
df
```

```python
df["Tanggal"].info()
```

```python
import plotly.express as px
fig = px.line(df,x="Tanggal", y='RR',title="Tahun 2016",markers=True)
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
