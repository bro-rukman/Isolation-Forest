```python
from sklearn.ensemble import IsolationForest
X = df[["RR"]]
iForest = IsolationForest()
# iForest = IsolationForest(n_estimators=10,max_samples="auto",contamination=0.05,bootstrap=False,n_jobs=-1,random_state=50)
preds = iForest.fit_predict(X)
df["scores"] = iForest.decision_function(X)
df["anomaly_label"] = preds
```
